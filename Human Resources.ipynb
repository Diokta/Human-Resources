## Data Preparation / Preprocessing

### Handling Missing or Invalid Values

```python
# Salin DataFrame asli agar tidak merusak data mentah
df_clean = df.copy()

# Daftar fitur yang mengandung nilai tidak valid
invalid_cols = ['Attrition']

# Hapus baris dengan nilai tidak valid
df_clean = df_clean.dropna(subset=['Attrition'])

# Verifikasi tidak ada lagi nilai NaN
print("\nCek ulang nilai NaN:")
print(df_clean.isnull().sum())
```

### Mengubah kolom categorical

```python
# Convert categorical columns to category type for plotting
categorical_cols = ['BusinessTravel', 'Department', 'EducationField', 'Gender',
                    'JobRole', 'MaritalStatus', 'OverTime', 'Over18']
for col in categorical_cols:
    df_clean[col] = df_clean[col].astype('category')

df_clean.info()
```

# Kelompokkan usia ke dalam bins

```python
df_clean['AgeGroup'] = pd.cut(df_clean['Age'], bins=[18, 25, 35, 45, 55, 65],
                        labels=['18–25', '26–35', '36–45', '46–55', '56–65'])
```

Dilakukan binning data Age untuk mengelompokkan data

## Data Visualization

## Attrition Rate

```python
total_employees = df_clean.shape[0]
total_attrition = df_clean[df_clean['Attrition'] == 1].shape[0]
attrition_rate = (total_attrition / total_employees) * 100

print(f"Total Employees: {total_employees}")
print(f"Total Attrition: {total_attrition}")
print(f"Attrition Rate: {attrition_rate:.2f}%")
```

## Attrition by Department

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df_clean, x='Department', hue='Attrition')
plt.title('Attrition by Department')
plt.show()
```

Terlihat bahwa departemen Research & Development memiliki jumlah karyawan terbanyak, termasuk jumlah karyawan yang keluar paling tinggi secara absolut. Namun, jika dilihat dari proporsinya, sebagian besar karyawan di departemen ini tetap bertahan, sehingga tingkat attrition relatif lebih rendah dibandingkan departemen lain. Sebaliknya, departemen Sales memiliki jumlah karyawan yang lebih sedikit, tetapi proporsi karyawan yang keluar terlihat cukup besar, yang mengindikasikan adanya masalah retensi yang lebih serius di departemen tersebut. Departemen Human Resources memiliki jumlah karyawan paling sedikit dan attrition yang sangat rendah, baik secara absolut maupun proporsional. Berdasarkan temuan ini, manajemen perlu memberi perhatian khusus pada departemen Sales dalam upaya mengurangi attrition rate, serta mempertimbangkan pendekatan retensi di departemen HR yang tampaknya lebih berhasil menjaga stabilitas tenaga kerja.

## Attrition Rate by length of work

```python
# Filter hanya karyawan yang keluar
resigned_employees = df_clean[df_clean['Attrition'] == 1]

# Hitung rata-rata lama bekerja
average_years_before_resign = resigned_employees['YearsAtCompany'].mean()

print(f"Rata-rata lama bekerja sebelum resign: {average_years_before_resign:.2f} tahun")

plt.figure(figsize=(8, 5))
sns.histplot(resigned_employees['YearsAtCompany'], bins=10, kde=True)
plt.title('Distribusi Lama Bekerja Sebelum Resign')
plt.xlabel('Tahun Bekerja')
plt.ylabel('Jumlah Karyawan')
plt.grid(True)
plt.show()
```

Terlihat bahwa sebagian besar karyawan yang resign melakukannya dalam rentang waktu 0 hingga 5 tahun pertama bekerja, dengan puncaknya terjadi pada tahun pertama atau kedua. Setelah 5 tahun, jumlah karyawan yang resign menurun drastis dan semakin sedikit seiring bertambahnya tahun kerja. Pola ini menunjukkan bahwa risiko attrition paling tinggi terjadi pada masa awal masa kerja, khususnya dalam 1–3 tahun pertama. Hal ini dapat disebabkan oleh kurangnya adaptasi, ketidakcocokan dengan lingkungan kerja, atau ketidakpuasan awal terhadap peran dan kompensasi.

## Attrition Rate by Age and Gender

```python
# Hitung attrition rate berdasarkan gender dan kelompok usia
pivot_table = df_clean.pivot_table(values='Attrition',
                             index='AgeGroup',
                             columns='Gender',
                             aggfunc='mean')

# Tampilkan heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table * 100, annot=True, fmt=".1f", cmap='coolwarm')
plt.title('Attrition Rate (%) Berdasarkan Usia dan Gender')
plt.ylabel('Kelompok Usia')
plt.xlabel('Gender')
plt.show()
```

Grafik “Attrition Rate (%) Berdasarkan Usia dan Gender” dalam bentuk heatmap ini menggambarkan persentase karyawan yang keluar (attrition rate) menurut kelompok usia dan jenis kelamin. Terlihat bahwa kelompok usia 18–25 tahun memiliki tingkat attrition tertinggi, khususnya pada karyawan perempuan dengan angka mencapai 51,6%, diikuti oleh laki-laki pada kelompok usia yang sama dengan 29,1%. Hal ini menunjukkan bahwa karyawan muda, terutama perempuan, cenderung lebih sering meninggalkan perusahaan.

Seiring bertambahnya usia, tingkat attrition cenderung menurun di hampir semua kelompok, dengan titik terendah untuk perempuan berada pada kelompok 46–55 tahun (4,2%) dan untuk laki-laki sedikit lebih tinggi di kelompok usia yang sama (15,5%). Pada kelompok usia 56–65 tahun, tingkat attrition kembali sedikit meningkat untuk kedua gender, kemungkinan karena mendekati usia pensiun.

Secara keseluruhan, grafik ini mengindikasikan bahwa usia muda dan jenis kelamin perempuan merupakan faktor risiko attrition yang tinggi. Perusahaan dapat memanfaatkan informasi ini untuk merancang kebijakan retensi yang lebih spesifik, seperti program pengembangan karier, mentorship, atau fleksibilitas kerja bagi karyawan muda, khususnya perempuan, guna meningkatkan retensi dalam kelompok paling rentan ini.

## Attrition Rate by Years At Company and Years In Current Role

```python
# Hitung rata-rata per status attrition
avg_stats = df.groupby('Attrition')[['YearsAtCompany', 'YearsInCurrentRole']].mean().rename(
    index={0: 'Tetap', 1: 'Keluar'}
)

print("Rata-rata Lama Bekerja dan Lama di Posisi Saat Ini")
print(avg_stats)

# Visualisasinya (bar chart)
import matplotlib.pyplot as plt

avg_stats.plot(kind='bar', figsize=(8, 6))
plt.title('Rata-rata YearsAtCompany dan YearsInCurrentRole (Keluar vs. Tetap)')
plt.ylabel('Tahun')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.legend(title='Metrik')
plt.tight_layout()
plt.show()
```

Terlihat bahwa karyawan yang tetap bekerja memiliki rata-rata lama bekerja di perusahaan sekitar 7,5 tahun dan lama menempati posisi saat ini sekitar 4,5 tahun. Sementara itu, karyawan yang keluar memiliki rata-rata lama bekerja lebih pendek, sekitar 5,2 tahun, dan masa jabatan di posisi terakhir hanya sekitar 3 tahun.

Perbedaan ini menunjukkan bahwa karyawan yang keluar cenderung memiliki pengalaman kerja yang lebih singkat, baik secara keseluruhan maupun dalam peran terakhir mereka. Hal ini dapat mengindikasikan bahwa kurangnya keterikatan terhadap perusahaan atau peran yang belum cukup berkembang bisa menjadi faktor penyebab attrition. Oleh karena itu, perusahaan disarankan untuk memperhatikan pengembangan karier dan rotasi jabatan bagi karyawan baru atau menengah agar mereka merasa lebih terlibat dan termotivasi untuk bertahan lebih lama.

## Attrition Rate by Years Since Last Promotion

```python
# Hitung attrition rate per nilai YearsSinceLastPromotion
promotion_group = df.groupby('YearsSinceLastPromotion')['Attrition'].mean() * 100

# Visualisasinya
plt.figure(figsize=(8, 5))
plt.plot(promotion_group.index, promotion_group.values, marker='o', linestyle='-')
plt.title('Attrition Rate Berdasarkan YearsSinceLastPromotion')
plt.xlabel('Tahun Sejak Promosi Terakhir')
plt.ylabel('Attrition Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Terlihat bahwa pada 0 hingga 3 tahun setelah promosi terakhir, tingkat attrition relatif stabil di kisaran 15–20%. Namun setelah itu, attrition rate menurun drastis, khususnya pada tahun ke-5 hingga ke-8, bahkan mencapai titik 0% pada tahun ke-8.

Menariknya, setelah tahun ke-8, attrition rate mulai berfluktuasi secara tajam — naik signifikan pada tahun ke-9 (sekitar 22%), turun menjadi 0% lagi pada tahun ke-10 dan 12, kemudian melonjak tinggi pada tahun ke-13 (mendekati 30%) dan kembali naik pada tahun ke-15. Fluktuasi ini kemungkinan terjadi karena jumlah karyawan yang belum dipromosikan dalam jangka waktu sangat lama relatif sedikit, sehingga setiap kasus keluar memberi dampak besar terhadap persentase.

Secara umum, grafik ini mengindikasikan bahwa karyawan yang tidak mendapatkan promosi dalam jangka waktu yang lama cenderung memiliki kemungkinan lebih besar untuk keluar, terutama setelah melewati masa 5–7 tahun tanpa promosi. Oleh karena itu, perusahaan perlu memperhatikan kebijakan promosi dan pengembangan karier untuk mencegah meningkatnya attrition di kalangan karyawan yang stagnan dalam perannya.

# Kesimpulan Analisis Attrition — PT Jaya Jaya Maju

Perusahaan **Jaya Jaya Maju** menghadapi tingkat attrition (karyawan keluar) yang cukup tinggi, yaitu lebih dari 10%. Berdasarkan analisis data visual yang dilakukan, berikut adalah rangkuman faktor-faktor utama penyebab attrition serta rekomendasi untuk mengatasinya.

---


## Faktor-Faktor Utama Penyebab Tingginya Attrition

1. **Peran dan Departemen Rentan Terhadap Attrition**
   - Jabatan seperti **Sales Representative** dan **Laboratory Technician** menunjukkan tingkat keluar yang tinggi secara proporsional.
   - Departemen **Sales** mengalami tingkat attrition lebih besar dibandingkan **R&D** atau **HR**.

2. **Pendapatan Bulanan Rendah Berkorelasi dengan Tingginya Attrition**
   - Karyawan yang keluar umumnya memiliki pendapatan bulanan yang lebih rendah dibandingkan karyawan yang bertahan.

3. **Keseimbangan Kerja dan Kehidupan (Work-Life Balance)**
   - Karyawan dengan skor *WorkLifeBalance* rendah (1 atau 2) lebih banyak yang keluar dibandingkan skor 3 atau 4.

4. **Beban Kerja Tinggi (Lembur) Menyumbang Tingkat Keluar**
   - Proporsi attrition lebih tinggi pada karyawan yang bekerja lembur (*OverTime = Yes*).

5. **Attrition Tinggi Terjadi di Awal Masa Kerja**
   - Mayoritas karyawan resign dalam 1–5 tahun pertama bekerja di perusahaan.

6. **Kelompok Usia Muda dan Perempuan Lebih Rentan**
   - Usia **18–25 tahun**, terutama perempuan, memiliki attrition rate tertinggi (hingga 51,6%).

7. **Kurangnya Promosi Meningkatkan Risiko Resign**
   - Attrition rate meningkat setelah 5–7 tahun tidak mendapat promosi.

8. **Lama Bekerja dan Lama di Posisi Saat Ini Berpengaruh**
   - Karyawan yang tetap memiliki masa kerja dan masa di posisi saat ini lebih panjang dibandingkan yang keluar.
